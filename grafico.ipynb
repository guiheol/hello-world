```python
# conteúdo de content/grafico.ipynb — primeira célula
import numpy as np
import matplotlib.pyplot as plt

# Dados: uma senóide e sua transformada "didática" (apenas exemplo)
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(3*x) * np.exp(-0.3*x)

plt.figure(figsize=(6,4))
plt.plot(x, y, label='y(x) = sin(3x) * e^{-0.3x}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exemplo de Gráfico com Matplotlib')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
